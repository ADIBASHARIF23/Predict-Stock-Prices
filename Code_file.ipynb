```python
data = df.copy()
data.dtypes
```

```python
#rolling & moving forward concepts
lag_features = ['High','Low','Volume','Turnover','Trades'] #contains all the lagging features
window1 =3
window2 =7
```

```python
#iterating all the lag_features to implement the rolling & moving forward concepts
for feature in lag_features:
    data[feature+'_rolling_mean_3']=data[feature].rolling(window=window1).mean()
    data[feature+'_rolling_mean_7']=data[feature].rolling(window=window2).mean()


for feature in lag_features:
    data[feature+'_rolling_std_3']=data[feature].rolling(window=window1).std()
    data[feature+'_rolling_std_7']=data[feature].rolling(window=window2).std()
```

```python
#creating an instance for all the independent features to pass through the model for training purpose
ind_features = ['High_rolling_mean_3', 'High_rolling_mean_7',
       'Low_rolling_mean_3', 'Low_rolling_mean_7', 'Volume_rolling_mean_3',
       'Volume_rolling_mean_7', 'Turnover_rolling_mean_3',
       'Turnover_rolling_mean_7', 'Trades_rolling_mean_3',
       'Trades_rolling_mean_7', 'High_rolling_std_3', 'High_rolling_std_7',
       'Low_rolling_std_3', 'Low_rolling_std_7', 'Volume_rolling_std_3',
       'Volume_rolling_std_7', 'Turnover_rolling_std_3',
       'Turnover_rolling_std_7', 'Trades_rolling_std_3',
       'Trades_rolling_std_7']
```

```python
#create new dataframes for training & test data
training_data = data[0:1800]
test_data = data[1800:]
```

**Apply auto_Arima on data**

```python
pip install pmdarima
```

```python
from pmdarima import auto_arima
```

```python
model= auto_arima(y=training_data['VWAP'],X=training_data[ind_features],trace=True)
```

```python
model.fit(y=training_data['VWAP'],X=training_data[ind_features])
```

```python
forecast = model.predict(n_periods=len(test_data),X=test_data[ind_features])
forecast
```

```python
#adding a column to the test_data
test_data['Forecast_ARIMA'] = forecast.tolist() #explicitely assign the values of farecast to this new column
test_data['Forecast_ARIMA'].values
```

```python
#visualize the diffrence between the actual and the predicted test data
test_data[['VWAP','Forecast_ARIMA']].plot(figsize=(14,7))
```

**Evaluate the model using both RMSE & MAE**

```python
from sklearn.metrics import mean_absolute_error,root_mean_squared_error
```

```python
print(mean_absolute_error(test_data['VWAP'],test_data['Forecast_ARIMA']))
print(root_mean_squared_error(test_data['VWAP'],test_data['Forecast_ARIMA']))
```
